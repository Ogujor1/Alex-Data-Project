<h5> Females are more responsive to properties than males.</h5>

```python
age_count = df["Age Range"].value_counts()
age_count
```

```python
age_df = pd.DataFrame({'age':age_count.index, 'age_count':age_count.values})
age_df
```

```python
plt.title('Age Group Response Chart', color='Red', fontsize=16)
plt.xlabel('Age Range', color='green')
plt.ylabel('Numbers', color='purple')
plt.bar(age_df['age'],age_df['age_count'])
plt.show()
```

<h5> Adults from the age 25 and above are most concern about properities.</h5>

```python
occ = df['Occupation'].value_counts()
occ
```

```python
occ_df = pd.DataFrame({'occupation':occ.index, 'values':occ.values})
occ_df
```

```python
plt.figure(figsize=(8,6))
plt.title('Occupational Chart', color='purple', fontsize=16)
y = occ_df["occupation"]
x = occ_df["values"]
explodes = [0.2,0,0,0]

plt.pie(x, labels=y, explode=explodes, shadow=True)
plt.legend(title='Occupations')
plt.show()
```

```python
reshaped_occ = occ_df.pivot(columns='occupation', values='values')
reshaped_occ.fillna(0, inplace=True)
reshaped_occ
```
